```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("Simple.csv")
df.head()
```

```python
df = df.dropna()
df
```

```python
step = 0.00001
to_bin = lambda x: np.floor(x / step) * step
df["lat"] = df.LAT.map(to_bin)
df["lon"] = df.LON.map(to_bin)
```

```python
df = df[["lat", "lon",  "CNTY_NAME", "CITY_TOWN_NAME"]]
df
```

```python
df = df.groupby(df.columns.tolist()).size().reset_index().\
    rename(columns={0:'num'})
df
```

```python
df.to_csv("SimpleDataSample.csv")
```
